```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
df_balance = pd.read_csv('fish_data.csv')
```

```python
x = df_balance.iloc[:,1:].values
y = df_balance.iloc[:,[0]].values
```

```python
y = y.reshape(len (y),)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
svm_model = SVC(kernel='linear', random_state=0)
svm_model.fit(x_train, y_train)

```

```python
y_pred = svm_model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

```

### kernel terbaik

```python
# List kernel yang akan diuji
kernels = ['linear', 'rbf', 'poly',  'sigmoid']
kernel_results = {}

# Loop melalui setiap kernel
for kernel in kernels:
    # Buat model dengan kernel tertentu
    svm = SVC(kernel=kernel, random_state=0)
    svm.fit(x_train, y_train)  # Latih model
    
    # Prediksi dan evaluasi
    y_pred = svm.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    
    # Simpan hasil
    kernel_results[kernel] = acc

# Cetak hasil untuk setiap kernel
for k, v in kernel_results.items():
    print(f"Kernel: {k}, Accuracy: {v}")

# Cari kernel terbaik
best_kernel = max(kernel_results, key=kernel_results.get)
print(f"Kernel terbaik: {best_kernel}")
```
